## Data Cleaning

Let's load the some more recent data for AAPL. Unfortunately, this data is not quite as *clean* as last year's data, so we'll need to do some data wrangling. The file we're looking to load is `AAPL_2026_raw.csv`.

````python
# New imports for a new notebook!
import pandas as pd
````

````python
# Loading the data, setting the index, and having a look
df = pd.read_csv("https://raw.githubusercontent.com/ImperialCollegeLondon/efds-ta-python/refs/heads/main/pyfi-data/AAPL_2026_raw.csv")
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
df.set_index("Date", inplace=True) # inplace automatically updates the dataframe
df
````

## Ordering, Columns, Duplicates

First let's start with sorting the index.

````python
# Checking for ascending order in the index
display(df.index.is_monotonic_increasing)

# Sorting the index in place
display(df.sort_index(inplace=True)) # Notice this returns nothing - but the sort works!

# Checking for ascending order in the index
df.index.is_monotonic_increasing
````

Next let's look at getting rid of unnecessary columns:

````python
# Dropping columns with the drop method
display(df.drop(columns=["Dividends", "Stock Splits"]))

# Checking to see if it worked...
display(df.columns)

# But make sure we update the dataframe to save our work!
display(df.drop(columns=["Dividends", "Stock Splits"], inplace=True))

# It worked!
df.columns
````

Now let's focus on duplicates:

````python
# Counting duplicates
display(df.duplicated().sum())

# Getting rid of duplicates
display(df.drop_duplicates())

# Counting duplicates again...
display(df.duplicated().sum())

# Getting rid of duplicates again... with inplace=True
display(df.drop_duplicates(inplace=True))

# Counting duplicates a final time...
df.duplicated().sum()
````

#### Tip: Method Chaining

**Method chaining** is a popular feature of pandas. It allows us to *chain* together several operations in a single line of code. For example, we can set the index, sort the data frame and drop duplicates and columns all at once. Notice we don't use `inplace` but rather re-assign to the original `df` variable.

```python
df = df.set_index("Date").sort_index().drop_duplicates().drop(columns=["Dividends", "Stock Splits"])
```

Now that we've identified our missing values, the big question is how to handle them. There are many approaches to this that will vary depending on the data and the further analysis you plan to carry out.

````python
# We can just drop rows that contains NaN in any column
display(df.dropna())

# Drop rows which contains NaN in a specific column
display(df.dropna(subset="Close"))

# We can fill in NaNs with the average of a column
display(df['Volume'].fillna(df['Volume'].mean()))

# We can interpolate (point on a line connecting the value of the days before and after)
display(df["Close"].interpolate(method="linear"))

# We can forward fill, and use the value from the day before
display(df["Close"].ffill())

# Notice that ffill() didn't work for our missing value in the first row
# A common approach is to chain ffill() with bfill() to handle it
df["Close"].ffill().bfill()
````

#### Advanced: Data Types

You may have noticed that the **Volume** column in the 2026 data frame is a `float64` instead of the `int64` *dtype* we had in the 2025 data frame. Missing values (NaN) are represented as a special case of floating point number, so all the values in **Volume** were automatically *upcast* to floats when the data was originally loaded from the CSV.

Ideally our columns should be of the *dtype* that most accurately represents them. This will improve performance when working with large data frames. Now that we've resolved our missing numbers, we can *cast* our trading volumes as integers.

````python
df["Volume"] = df["Volume"].astype("int64")
df.info()
````
